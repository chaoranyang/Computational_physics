# 1.eqution

$$
m\ddot{\vec{r}}=\frac{-G_NM_{sun}m}{r^2}\frac{\vec{r}}{r}=\frac{-c^2}{2}\left( \frac{2G_NM_{sun}m}{c^2} \right) \frac{1}{r^2}\frac{\vec{r}}{r}
$$


$$
\ddot{\vec{r}}=-\frac{c^2}{2}\left( r_s \right) \frac{1}{r^2}\frac{\vec{r}}{r}
$$
where $r_s=\frac{2G_NM_{sun}}{c^2}$

# 2. Orbit of Mercury: Newton

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R0 = 1.0  
T0 = 1.0  
cbam = 0.99  

x0 = 0.0
y0 = 5.5 * R0
vx0 = 0.53 * R0/T0
vy0 = 0.0

myed = 88.0  
tmy= 5  
totT = tmy* myed * T0 

def acc(x, y):
    r = np.sqrt(x**2 + y**2)
    r3 = r**3
    ax = -cbam * x / r3
    ay = -cbam * y / r3
    return ax, ay
```

```python
def euler_method(dt, totT):
    stepn = int(totT / dt)
    x = np.zeros(stepn)
    y = np.zeros(stepn)
    vx = np.zeros(stepn)
    vy = np.zeros(stepn)
    x[0] = x0
    y[0] = y0
    vx[0] = vx0
    vy[0] = vy0
    
    for i in range(stepn - 1):
        ax, ay = acc(x[i], y[i])
        
        vx[i+1] = vx[i] + ax * dt
        vy[i+1] = vy[i] + ay * dt
        
        x[i+1] = x[i] + vx[i] * dt
        y[i+1] = y[i] + vy[i] * dt
    
    return x, y, vx, vy
```

```python
def leap_frog_method(dt, totT):
    stepn = int(totT / dt)
    
    x = np.zeros(stepn)
    y = np.zeros(stepn)
    vx = np.zeros(stepn)
    vy = np.zeros(stepn)
    
    x[0] = x0
    y[0] = y0
    vx[0] = vx0
    vy[0] = vy0
    
    ax, ay = acc(x[0], y[0])
    vx_h = vx[0] + 0.5 * ax * dt
    vy_h = vy[0] + 0.5 * ay * dt
    
    for i in range(stepn - 1):
        x[i+1] = x[i] + vx_h * dt
        y[i+1] = y[i] + vy_h * dt
        ax, ay = acc(x[i+1], y[i+1])
        
        vx_h = vx_h + ax * dt
        vy_h = vy_h + ay * dt
        
        vx[i+1] = vx_h - 0.5 * ax * dt
        vy[i+1] = vy_h - 0.5 * ay * dt
    
    return x, y, vx, vy
```

```python
def rk4_method(dt, totT):
    stepn = int(totT / dt)
    
    x = np.zeros(stepn)
    y = np.zeros(stepn)
    vx = np.zeros(stepn)
    vy = np.zeros(stepn)
    
    x[0] = x0
    y[0] = y0
    vx[0] = vx0
    vy[0] = vy0
    
    for i in range(stepn - 1):
        k1x = vx[i]
        k1y = vy[i]
        a1x, a1y = acc(x[i], y[i])
        k1vx = a1x
        k1vy = a1y
        
        k2x = vx[i] + 0.5 * dt * k1vx
        k2y = vy[i] + 0.5 * dt * k1vy
        a2x, a2y = acc(x[i] + 0.5 * dt * k1x, y[i] + 0.5 * dt * k1y)
        k2vx = a2x
        k2vy = a2y
        
        k3x = vx[i] + 0.5 * dt * k2vx
        k3y = vy[i] + 0.5 * dt * k2vy
        a3x, a3y = acc(x[i] + 0.5 * dt * k2x, y[i] + 0.5 * dt * k2y)
        k3vx = a3x
        k3vy = a3y
        
        k4x = vx[i] + dt * k3vx
        k4y = vy[i] + dt * k3vy
        a4x, a4y = acc(x[i] + dt * k3x, y[i] + dt * k3y)
        k4vx = a4x
        k4vy = a4y
        
        x[i+1] = x[i] + (dt/6) * (k1x + 2*k2x + 2*k3x + k4x)
        y[i+1] = y[i] + (dt/6) * (k1y + 2*k2y + 2*k3y + k4y)
        vx[i+1] = vx[i] + (dt/6) * (k1vx + 2*k2vx + 2*k3vx + k4vx)
        vy[i+1] = vy[i] + (dt/6) * (k1vy + 2*k2vy + 2*k3vy + k4vy)
    
    return x, y, vx, vy
```

```python
plt.figure(figsize=(15, 10))

colors = ['blue', 'red', 'green']
linestyles = ['-', '--', '-.']

dt_euler = [3e-4, 3e-1, 1.0]
for i, dt in enumerate(dt_euler):
    x, y, vx, vy = euler_method(dt, totT)
    plt.plot(x, y, color=colors[0], linestyle=linestyles[i], 
             label=f'Euler, Δt = {dt}', linewidth=1.5)

dt_leapfrog = [3e-1, 1.0]
for i, dt in enumerate(dt_leapfrog):
    x, y, vx, vy = leap_frog_method(dt, totT)
    plt.plot(x, y, color=colors[1], linestyle=linestyles[i], 
             label=f'Leap-Frog, Δt = {dt}', linewidth=1.5)

dt_rk4 = [3e-1, 1.0]
for i, dt in enumerate(dt_rk4):
    x, y, vx, vy = rk4_method(dt, totT)
    plt.plot(x, y, color=colors[2], linestyle=linestyles[i], 
             label=f'RK4, Δt = {dt}', linewidth=1.5)

plt.plot(0, 0, 'ro', markersize=15, label='Sun', markeredgecolor='black')

plt.xlabel('x (R₀)')
plt.ylabel('y (R₀)')
plt.title('Orbit of Mercury: Different Methods (5 Mercury Years)')
plt.axis('equal')
plt.grid(True, alpha=0.3)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
methods = [
    ('Euler Method', euler_method, [3e-4, 3e-1, 1.0]),
    ('Leap-Frog Method', leap_frog_method, [3e-1, 1.0]),
    ('4th-order Runge-Kutta', rk4_method, [3e-1, 1.0])
]

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for idx, (method_name, method_func, dt_list) in enumerate(methods):
    ax = axes[idx]
    
    for i, dt in enumerate(dt_list):
        x, y, vx, vy = method_func(dt, totT)
        ax.plot(x, y, linestyle=linestyles[i], linewidth=1.5, label=f'Δt = {dt}')
    
    ax.plot(0, 0, 'ro', markersize=10, label='Sun')
    ax.set_xlabel('x (R₀)')
    ax.set_ylabel('y (R₀)')
    ax.set_title(f'{method_name}')
    ax.axis('equal')
    ax.grid(True, alpha=0.3)
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
def toten(x, y, vx, vy):
    r = np.sqrt(x**2 + y**2)
    Ek = 0.5 * (vx**2 + vy**2)
    Ep = -cbam / r
    return Ek + Ep

plt.figure(figsize=(12, 8))


dt_energy = 3e-1


x_e, y_e, vx_e, vy_e = euler_method(dt_energy, totT)
stepn = len(x_e)
time = np.arange(stepn) * dt_energy
energies_euler = np.zeros(stepn)
for i in range(stepn):
    energies_euler[i] = toten(x_e[i], y_e[i], vx_e[i], vy_e[i])

    
x_lf, y_lf, vx_lf, vy_lf = leap_frog_method(dt_energy, totT)
energies_leapfrog = np.zeros(stepn)
for i in range(stepn):
    energies_leapfrog[i] = toten(x_lf[i], y_lf[i], vx_lf[i], vy_lf[i])

    
x_rk, y_rk, vx_rk, vy_rk = rk4_method(dt_energy, totT)
energies_rk4 = np.zeros(stepn)
for i in range(stepn):
    energies_rk4[i] = toten(x_rk[i], y_rk[i], vx_rk[i], vy_rk[i])
```

```python
plt.plot(time, energies_euler, 'r-', linewidth=1.5, label='Euler Method')
plt.plot(time, energies_leapfrog, 'g-', linewidth=1.5, label='Leap-Frog Method')
plt.plot(time, energies_rk4, 'b-', linewidth=1.5, label='4th-order Runge-Kutta')

plt.xlabel('Time (Earth days)')
plt.ylabel('Total Energy')
plt.title('Energy Conservation Comparison (Δt = 0.3)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
